```python
# droping irrelvent columns
reset_data = pol_ins.drop(["Domain Code", "Domain", "Area Code", "Element Code", "Element", "Item Code", "Item", 
"Year Code", "Unit", "Flag", "Note"], axis=1 )
reset_data.head()
```

```python
import plotly.express as px

# df = px.data.gapminder().query("country=='Canada'")
fig = px.line(reset_data, x="Year", y="Value", title='Political stability and absence of violence/terrorism in Pakistan')

fig.show()
```

#### From above visualization we see that with the passage of time stability in politics increasing and violence and terrorism also increasing.
